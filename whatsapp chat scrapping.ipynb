```python
import os
import pickle
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from time import sleep
from selenium.webdriver.chrome.service import Service
import chromedriver_autoinstaller


def get_visible_messages(driver):
    print('>>> getting visible messages')
    messages = set()
    elements = driver.find_elements_by_xpath("//div[@class='_amjy _amjz _amjw']")
    for e in elements:
        messages.add(e.get_attribute('data-pre-plain-text') + e.text)
    return messages

def main():
    # Path to the ChromeDriver executable
    #CHROMEDRIVER_PATH = r'C:\Users\Muhammad Usama\Desktop\drivers\alpha\chromedriver-win64\chromedriver.exe'
    chromedriver_autoinstaller.install()

    # Check if the specified path is a valid file
    #if not os.path.exists(CHROMEDRIVER_PATH):
        #print("Error: The specified path to ChromeDriver does not exist or is invalid.")
        #return

    options = Options()
    options.add_argument('user-data-dir=./User_Data')
    service = Service()
    options = Options()
    options.add_argument('user-data-dir=./User_Data')
    options.add_argument("--headless")
    options.add_argument("--disable-gpu")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--enable-logging")
    options.add_argument("--v=1")
    service = Service(log_path='chromedriver.log')




    try:
        # Use ChromeDriver directly without passing executable_path
        #driver = webdriver.Chrome(CHROMEDRIVER_PATH, options=options)
        driver = webdriver.Chrome(service=service, options=options)

        driver.get('https://web.whatsapp.com/')
        input('Press enter after scanning QR code or after the page has fully loaded\n')

        print('>>> getting contact list')
        contacts_sel = driver.find_elements_by_class_name('_x10l6tqk _xh8yej3 _x1g42fcv')
        contacts = set([j.text for j in contacts_sel])

        conversations = []
        for contact in contacts:
            user = driver.find_element_by_xpath('//span[contains(@title, "{}")]'.format(contact))
            user.click()
            sleep(2)  # Allow time for the chat to open
            messages = get_visible_messages(driver)
            conversations.append(messages)
            filename = f'collected_data/conversations/{contact}.json'
            os.makedirs(os.path.dirname(filename), exist_ok=True)
            with open(filename, 'wb') as fp:
                pickle.dump(messages, fp)

        print(f"{len(contacts)} contacts retrieved")
        print(f"{len(conversations)} conversations retrieved")
        filename = 'collected_data/all.json'
        os.makedirs(os.path.dirname(filename), exist_ok=True)
        with open(filename, 'wb') as fp:
            pickle.dump(conversations, fp)
    except Exception as e:
        print(e)
    finally:
        try:
            driver.quit()
        except NameError:
            pass

if __name__ == "__main__":
    main()
```
